### Orthogonal Vectors

```python
import numpy as np
```

Baris ini berfungsi untuk mengimpor library NumPy, yang merupakan library standar di Python untuk pengolahan matriks dan komputasi numerik

```python
i = np.array([1, 0])
i
```

 Di sini kita membuat dua buah vektor dalam ruang 2 dimensi  menggunakan fungsi np.array()

Vektor i (1, 0) dan j (0, 1) adalah Basis Vectors standar.

```python
j = np.array([0, 1])
j
```

Baris ini berfungsi untuk membuat sebuah objek array NumPy yang merepresentasikan vektor secara matematis.
Angka 0 adalah komponen pada sumbu x, dan 1 adalah komponen pada sumbu y

```python
np.dot(i, j)
```

Fungsi np.dot(i, j) digunakan untuk menghitung Dot Product (perkalian titik) antara vektor $i$ dan $j$. Hasil outputnya adalah 0
