```
#buat variabel x dan simpan suatu nilai skalar lalu tampilkan tipe data x
x = 13
print(x)
type(x)
```

```
#buat variabel y dan simpan suatu nilai skalar lalu tampilkan data y
y=31
print(y)
type(y)
```

```
#jumlahkan x+y
sum_py=x+y
print(sum_py)
```

```
#buat variabel dalam x_float dan simpan suatu nilai desimal dan jumlahkan x_float+y, serta tampilkan tipe data x_float
x_float=1.3
print(x_float)
type(x_float)
z = x_float + y
print(z)

```

```
import tensorflow as tf
```

```
#gunakan soal nomor 1 tapi dengan tensorflow
```

```
x_tf = tf.Variable(13.0, dtype_=tf.float32)
print(x_tf)
```

```
y_tf = tf.Variable(31.0, dtype_=tf.float32)
print(y_tf)
```

```
#jumlahkan x dan y
p = x_tf + y_tf
print(p)
```

```
#buat variable x_float
x_float_tf = tf.Variable(1.3, dtype_=tf.float32)
print(x_float_tf)
f = x_float_tf + y_tf
print(f)
```
